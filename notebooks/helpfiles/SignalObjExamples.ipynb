````python
from pathlib import Path
import sys
import importlib
import json

def find_repo_root(start: Path) -> Path:
    cur = start.resolve()
    for p in [cur, *cur.parents]:
        if (p / '.git').exists() and (p / 'helpfiles').exists():
            return p
    raise RuntimeError('Could not find repository root')

repo_root = find_repo_root(Path.cwd())
py_root = repo_root / 'python'
if str(py_root) not in sys.path:
    sys.path.insert(0, str(py_root))
print('repo_root =', repo_root)
````

````python
import re
import html as _html

html_path = repo_root / 'helpfiles/SignalObjExamples.html'
if not html_path.exists():
    print('HTML reference missing:', html_path)
else:
    html_text = html_path.read_text(encoding='utf-8', errors='ignore')
    title_match = re.search(r'<title>(.*?)</title>', html_text, flags=re.I | re.S)
    title = _html.unescape(re.sub(r'<[^>]+>', '', title_match.group(1))).strip() if title_match else html_path.stem
    sections = [
        _html.unescape(re.sub(r'<[^>]+>', '', s)).strip()
        for s in re.findall(r'<h2[^>]*>(.*?)</h2>', html_text, flags=re.I | re.S)
    ]
    sections = [s for s in sections if s]
    figs = sorted(set(re.findall(r'src="([^"]+_\d+\.png)"', html_text, flags=re.I)))
    print('HTML title:', title)
    print('Section count:', len(sections))
    for s in sections:
        print(' -', s)
    print('Figure refs:', len(figs))
    for f in figs[:20]:
        print(' -', f)
````
