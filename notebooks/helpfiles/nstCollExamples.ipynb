```python
from pathlib import Path
import json
import sys

def find_repo_root(start: Path) -> Path:
    cur = start.resolve()
    for p in [cur, *cur.parents]:
        if (p / '.git').exists() and (p / 'nstat').exists() and (p / 'helpfiles').exists():
            return p
    raise RuntimeError('Could not find nSTAT repo root from notebook cwd')

repo_root = find_repo_root(Path.cwd())
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))
print('repo_root =', repo_root)
```

```python
from examples.help_topics.nstCollExamples import run
expected_figures = 3
figure_dir = repo_root / 'reports' / 'figures' / 'notebooks' / 'nstCollExamples'
out = run(repo_root=repo_root, figure_dir=figure_dir, render_figures=True)
print(json.dumps(out, indent=2, default=str))
```

```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for fig_path in out.get('figures', []):
    p = Path(fig_path)
    img = mpimg.imread(p)
    h, w = img.shape[:2]
    fig = plt.figure(figsize=(max(w / 160.0, 2.0), max(h / 160.0, 2.0)), dpi=160)
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray' if img.ndim == 2 else None)
    ax.axis('off')
    ax.set_title(p.name)
    plt.show()
    plt.close(fig)
```

```python
assert out.get('topic') == 'nstCollExamples'
assert out.get('figure_contract_expected') == expected_figures
assert out.get('figure_count') == expected_figures, out
for fig_path in out.get('figures', []):
    assert Path(fig_path).exists(), fig_path
print('Notebook execution + figure contract: PASS')
```
